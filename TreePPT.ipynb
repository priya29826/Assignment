Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 :

       1

     / Â  \

   2Â  Â  Â  3

  / \Â  Â  / \

4 Â  5Â  6 Â  7

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.

Input2 :

 Â  Â  Â  1

     /Â  Â  \

  -2Â  Â  Â  3

  / \Â  Â  /Â  \

4 Â  5Â  -6 Â  2

Output2 : 7

Subtree with largest sum is :

 -2

 / \

4 Â  5

Also, entire tree sum is also 7.

</aside>

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

max_sum = float('-inf')

def maxSubtreeSum(root):
    global max_sum
    if root is None:
        return 0

    # Calculate the sum of the subtree rooted at the current node
    subtree_sum = root.val + maxSubtreeSum(root.left) + maxSubtreeSum(root.right)

    # Update the maximum sum if the current subtree sum is greater
    max_sum = max(max_sum, subtree_sum)

    return subtree_sum

# Example usage:
# Create the binary tree for Input1
root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)
root1.left.left = TreeNode(4)
root1.left.right = TreeNode(5)
root1.right.left = TreeNode(6)
root1.right.right = TreeNode(7)

print("Input1:")
# Expected Output: 28
maxSubtreeSum(root1)
print(max_sum)

# Create the binary tree for Input2
root2 = TreeNode(1)
root2.left = TreeNode(-2)
root2.right = TreeNode(3)
root2.left.left = TreeNode(4)
root2.left.right = TreeNode(5)
root2.right.left = TreeNode(-6)
root2.right.right = TreeNode(2)

print("Input2:")
# Expected Output: 7
max_sum = float('-inf')
maxSubtreeSum(root2)
print(max_sum)
```

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

</aside>

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def constructBST(arr):
    if not arr:
        return None

    queue = []
    root = TreeNode(arr.pop(0))
    queue.append(root)

    while queue:
        curr_node = queue.pop(0)
        curr_val = curr_node.val

        if arr:
            left_val = arr.pop(0)
            if left_val is not None:
                curr_node.left = TreeNode(left_val)
                queue.append(curr_node.left)

        if arr:
            right_val = arr.pop(0)
            if right_val is not None:
                curr_node.right = TreeNode(right_val)
                queue.append(curr_node.right)

    return root


arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(arr)

# Printing the BST using inorder traversal to verify the output
def inorderTraversal(node):
    if node is None:
        return
    inorderTraversal(node.left)
    print(node.val, end=' ')
    inorderTraversal(node.right)

inorderTraversal(root)
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

</aside>

```python
def isLevelOrderBST(arr):
    index = 0
    n = len(arr)

    while index < n:
        root = arr[index]
        left = []
        right = []

        index += 1
        while index < n and arr[index] < root:
            left.append(arr[index])
            index += 1

        while index < n and arr[index] > root:
            right.append(arr[index])
            index += 1

        # Check if any element in the right sublist is smaller than root
        if any(elem < root for elem in right):
            return False

    return True

# Example usage:
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
arr2 = [11, 6, 13, 5, 12, 10]

print(isLevelOrderBST(arr1))
print(isLevelOrderBST(arr2))
```
